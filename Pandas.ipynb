### loc[] & iloc[] -

```python
# using these 'loc[]' and 'iloc[]' we can fetch the rows and coulmns
```

```python
# loc[] - To fetch the rows 

df.loc[3]
```

```python
df.loc[0:4]
```

```python
df.loc[df.Country == 'India']
```

```python
df.loc[df.Rank < 20]
```

```python
# iloc[] - we can fetch both rows and columns using this method

df.iloc[3] # fetch row
```

```python
df.iloc[3, 3]  # fetch the value from row '4' and column '4'.
```

```python
df.iloc[[0, 2, 4, 7]] # fetch the any index, just sepearate it with comma
```

```python
df.iloc[1: 5, 2: 5]  # selecting rows from 1 to 4 and columns from 2 to 4 
```
