### Cleaning data using pandas -

```python
# Data cleaning is one of the most common tasks in data science.

import pandas as pd
Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv", usecols=['name','sex','age','fare'])
Data.head(10)
```

```python
Data.isnull().sum()
```

```python
# Dealing with missing data technique - 

# 1. Dropping missing values - this method will remove all the rows which are null/empty values.
# This method is not an effecient way to deal with data. Here data loss could be occur.

data = Data.dropna()
data.isnull().sum()
```

```python
Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv")

Data.dropna(inplace = True, axis=0) # here axis = 0 is rows and axis = 1 is columns
print(Data.head())
print(Data.isnull().sum())
```

```python
# 2. Replacing missing values - 

Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv", usecols=['age','fare','survived'])

# replacing the missing value using mean()
mean_value=Data['age'].mean()
print(mean_value)

data = Data.fillna(mean_value)
data.tail(10)
```

```python
# replacing the missing value using median()

Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv", usecols=['age','fare','survived'])

median_value= Data['age'].median()
print(median_value)

data1 = Data.fillna(median_value)
data1.tail(10)
```

```python
# value_counts() - divides into categories and counts how many elements are there in that category

Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv")
Data['sex'].value_counts()
```

```python
# groupby() - aggregating the data

Data = pd.read_csv(r"S:\Datasets\Sample dataset\titanic3.csv", usecols=['age','fare','survived','ticket'])

mean = Data['age'].mean()
print(mean)

Data.fillna(mean)

Data.groupby('ticket').mean().head()  # need to refer more about this
```
