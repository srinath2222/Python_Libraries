### Matplotlib pyplot - 

```python
# Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

import matplotlib.pyplot as plt
```

```python
# Now we will see most used plots in matplotlib are:

# 1. Line plot - 

import matplotlib.pyplot as plt

x=[1,2,4,5,6]
y=[1,3,5,7,8]

plt.plot(x,y)
plt.show
```

```python
# 2. Bar plot -
plt.bar(x,y)
plt.show()
```

```python
# 3. Histogram plot -
plt.hist(y)
plt.show
```

```python
plt.hist(x)
plt.hist(y)
plt.show
```

```python
# 4. scatter plot - 
plt.scatter(x,y)
plt.show
```

```python
# 5. pie chart -

category = ['Indog','Husky','Rottweiler','Labrador']
count = [133340,64677,37474,64747]

plt.pie(count, labels=category,autopct = '%1.1f%%',startangle = 90)
plt.title('Dog Breeds')
plt.show
```

```python
# 6. Area plot -

x=[1,2,4,5,6,7,8,15]
y1,y2=[1,3,5,7,8,10,12,13],[2,3,4,6,7,11,14,15]

plt.fill_between(x,y1,y2,color='pink',alpha=0.4,label='Area 1-2')
plt.plot(x,y1, label='Line 1', marker = '*')
plt.plot(x,y2, label='Line 2', marker = 'o')

plt.xlabel('x-axis'),plt.ylabel('y-axis'),plt.title('Area Chart')

plt.legend(),plt.show()
```

### Introduction to Figures -

```python
# constrained layout automatically adjusts subplots.

fig = plt.figure(figsize=(4,3),facecolor='deepskyblue',layout='constrained')
fig.suptitle('figure')
ax=fig.add_subplot()
ax.set_title('Axes',loc='left',fontstyle='oblique',fontsize='medium')

x=[0.1,0.2,0.4,0.5,0.6]
y=[0.1,0.3,0.5,0.7,0.8]

plt.plot(x,y)
plt.show
```

### subplot() -

```python
# To create multiple plots in a single figure, making comparing and analyzing different datasets or variables easier.

fig = plt.figure()
ax1 = fig.add_subplot(111)
ax1.plot(1,2,3)
ax2 = fig.add_subplot(222 , facecolor='pink')
ax2.plot(1,2,3)

x=[1,2,3,4,5,6]
y=[1,3,5,7,8,9]

plt.plot(x,y)
plt.show
```

```python
# Another example for subplot

import matplotlib.pyplot as plt
import numpy as np

figg, a = plt.subplots(2,2) # (rows,columns) - assume each subplot as each element in a matrix.

x = np.arange(1,5) 

a[0][0].plot(x , x*x)
a[0][0].set_title('Square')

a[0][1].plot(x , np.sqrt(x))
a[0][1].set_title('Square Root')

a[1][0].plot(x , np.exp(x))
a[1][0].set_title('Exp')

a[1][1].plot(x , np.log10(x))
a[1][1].set_title('Log')

plt.show
```
